```python
# Sort DataFrame based on another list
df = pd.DataFrame([["A", 1], ["B", 2],
                   ["C", 3], ["D", 4]], columns=["col1", "col2"])

sort_list = ["C", "A", "D", "B"]
df['col1'] = pd.Categorical(df['col1'], categories=sort_list, ordered=True)
new_df = df.sort_values(by='col1')
new_df
```

```python
# Swap two rows of a dataframe
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]], columns=['A', 'B'])
# Indices of rows to swap
row1 = 0
row2 = 2
# Swap the rows using iloc and tuple unpacking
df.iloc[[row1, row2]] = df.iloc[[row2, row1]].values
df
```
